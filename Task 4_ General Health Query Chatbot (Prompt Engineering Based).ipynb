### Inference client

```python
import os
from huggingface_hub import InferenceClient
from google.colab import userdata

def get_inference_client():
    # from huggingface_hub import login
    # from google.colab import userdata
    # login(userdata.get('HF_TOKEN'), add_to_git_credential=False)
    return InferenceClient(
        api_key=userdata.get('HF_TOKEN'),
    )
```

### Local pipeline

```python
from transformers import pipeline

def get_pipe():
    return pipeline("text-generation", model="Qwen/Qwen2.5-1.5B-Instruct", model_kwargs={"dtype": "auto", "device_map": "auto"})
```

### Send message to LLM either locally or via inference client

```python
messages = []

pipe = None
client = None

def send(msg: str, local = False):
    global pipe
    global client

    messages.append({"role": "user", "content": msg})

    if local:
        if pipe is None:
            pipe = get_pipe()

        response = pipe(messages, do_sample=True, max_new_tokens=256, temperature=0.6)
        content = response[0]["generated_text"][-1]["content"]
    else:
        if client is None:
            client = get_inference_client()

        completion = client.chat.completions.create(
            model="meta-llama/Llama-3.1-8B-Instruct:novita",
            messages=messages
        )
        content = completion.choices[0].message.content

    messages.append({"role": "assistant", "content": content})
    print("\nAI: " + content)
    print('\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
```

```python
system_msg = "Act like a helpful medical assistant. Provide general information. Safety Filter: Never prescribe medication. Always include a disclaimer that this is not professional medical advice. Be concise, and do not repeat instructions."
```

### Main loop

```python
def run():
    local = int(input("Enter:\n1 for local (use GPU runtime)\n0 for inference API (needs HF_TOKEN in colab secrets):\n"))
    if local != 1 and local != 0:
        raise Exception("Invalid input")
    local = (local == 1)

    global messages;
    messages = [
        {"role": "system", "content": system_msg},
    ]
    print('\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
    print("Welcome to Medical Assistant! Type below.")
    print("Enter 'quit' to exit.")
    print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')

    while True:
        msg = input("\nYou: ")
        if not msg.strip():
            continue

        if msg == 'quit':
            print("Take care!\n")
            break

        send(msg, local=local)
```

### Output if using inference

```python
run()
```

### Output if using local

```python
run()
```
