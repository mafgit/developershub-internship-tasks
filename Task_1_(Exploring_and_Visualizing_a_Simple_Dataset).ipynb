# Task 1: Exploring and Visualizing a Simple Dataset

```python
import seaborn as sns

df = sns.load_dataset('iris')

df.head()
```

### Histograms

```python
import matplotlib.pyplot as plt

plt.title("Correlation b/w Sepal Length & Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.scatter(df["sepal_length"], df["petal_length"])
plt.show()

plt.title("Correlation b/w Sepal Width & Petal Width")
plt.xlabel("Sepal Width")
plt.ylabel("Petal Width")
plt.scatter(df["sepal_width"], df["petal_width"])
plt.show()
```

```python
plt.title("Sepal Lengths")
plt.hist(df["sepal_length"], bins=15, edgecolor='black')
plt.show()

plt.title("Sepal Widths")
plt.hist(df["sepal_width"], bins=15, edgecolor='black')
plt.show()

plt.title("Petal Lengths")
plt.hist(df["petal_length"], bins=15, edgecolor='black')
plt.show()

plt.title("Petal Widths")
plt.hist(df["petal_width"], bins=15, edgecolor='black')
plt.show()
```

### Box Plots

```python
plt.title("Sepal Length Box Plot")
plt.boxplot(df["sepal_length"])
plt.show()

plt.title("Sepal Width Box Plot")
plt.boxplot(df["sepal_width"])
plt.show()

plt.title("Petal Length Box Plot")
plt.boxplot(df["petal_length"])
plt.show()

plt.title("Petal Width Box Plot")
plt.boxplot(df["petal_width"])
plt.show()
```
